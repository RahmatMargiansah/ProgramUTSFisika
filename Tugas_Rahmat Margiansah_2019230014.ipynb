#Rumus yang dibutuhkan

```python
# Fungsi mencari resistor pengganti/total untuk rangkaian seri
def resistorSeri(arrResistor):
  rTotal = 0
  for i in arrResistor:
    rTotal += i
  return rTotal

# Fungsi mencari resistor pengganti/total untuk rangkaian pararel
def resistorParalel(arrResistor):
  rTotal = 0
  for i in arrResistor:
    rTotal += 1/i
  return 1/rTotal

# Fungsi mencari kapasitor pengganti/total untuk rangkaian seri
def kapasitorSeri(arrKapasitor):
  cTotal = 0
  for i in arrKapasitor:
    cTotal += 1/i
  return 1/cTotal

# Fungsi mencari kapasitor pengganti/total untuk rangkaian pararel
def kapasitorParalel(arrKapasitor):
  cTotal = 0
  for i in arrKapasitor:
    cTotal += i
  return cTotal
```

```python
# Nilai tegangan
v = 15

# Nilai masing-masing resistor
r1 = 3
r2 = 6
r3 = 1
r4 = 1
```

#####a. R Pengganti

```python
# Rp1 pengganti R1 dan R2
rP1 = resistorParalel([r1, r2])

# Rp2 pengganti R3 dan R4
rP2 = resistorSeri([r3, r4])

# Rp pengganti Rp1 dan Rp2
rP = resistorParalel([rP1, rP2])

print("R penggantinya adalah ", rP, " Ohm",)
```

#####b. Arus yang melewati tegangan

```python
# I = V/Rp
i = v/rP

print("Arusnya adalah ", i, " Ampere")
```

#####c. Arus yang melewati masing-masing resistor

```python
# karena Rp1 merupakan rangkaian paralel, maka:
# I pada R1 = V/R1
i_r1 = v/r1
# I pada R2 = V/R2
i_r2 = v/r2

# karena Rp2 merupakan rangkaian seri, maka:
# I2 = I pada R3 = I pada R4
# I2 = V/Rp2
i2 = v/rP2
print("I pada R1 adalah ", i_r1, "Ampere")
print("I pada R2 adalah ", i_r2, "Ampere")
print("I pada R3 adalah ", i2, "Ampere")
print("I pada R4 adalah ", i2, "Ampere")
```

#Soal 4
Sebuah peralatan listrik menarik arus 5.5A ketika dihubungkan ke tegangan 110V.
<ol type='a'>
<li>Jika tegangan turun 10 persen, berapakah arus yang mengalir sekarang?</li>
<li>Jika hambatan berkurang 10 persen, berapa arus yang ditarik alat tersebut saat dihubungkan ke tegangan?</li>
</ol>

```python
# Nilai arus
i = 5.5

# Nilai tegangan
v = 110

# Nilai hambatan
r = v/i
```

#####a. Jika tegangan turun 10 persen, berapakah arus yang mengalir sekarang?

```python
# tegangan menjadi 90%
vTemp = v * 0.9

# Nilai arus baru
iTemp = vTemp/r

print("Nilai arus yang baru adalah ", iTemp, " Ampere")
```

#####b. Jika hambatan berkurang 10 persen, berapa arus yang ditarik alat tersebut saat dihubungkan ke tegangan?

```python
# Hambatan menjadi 90%
rTemp = r * 0.9

# Nilai arus baru
iTemp = v/rTemp

print("Nilai arus yang baru adalah ", iTemp, " Ampere")
```

#Soal 5
Sebuah solenoid dengan panjangnya 4.0 cm dan luas penampang 0.5 cm<sup>2</sup> memiliki 200 lilitan
<ol type='a'>
<li>Jika di rongga solenoid hanya berisi udara, berapa induktansi diri solenoid tersebut?</li>
<li>Berapa induktansi solenoid jika rongga solenoid berisi teras besi dengan $\mu$ = 4000? </li>
</ol>

```python
import math

# Nilai panjang diubah ke meter 0.04 m
l = 0.04

# Luas penampang diubah ke m^2 5x10^5 atau 0.00005
A = 0.00005

# Jumlah lilitan
N = 200

# Permeabilitas udara
mu0 = 4 * math.pi * 10**-7
```

#####a. Jika di rongga solenoid hanya berisi udara, berapa induktansi diri solenoid tersebut?

```python
L = N**2 * mu0 * A / l

print("Induktansi diri dari solenoidnya adalah ", L, " H")
```

#####b. Berapa induktansi solenoid jika rongga solenoid berisi teras besi dengan  $\mu$  = 4000?

```python
# Rongga solenoid berisi teras besi dengan mu = 4000
mu = 4000

L = N**2 * mu * mu0 * A / l
print("Induktansi diri dari solenoidnya adalah ", L, " H")
```
